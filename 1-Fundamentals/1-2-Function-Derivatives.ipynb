# Functions and their derivatives



```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp
from sympy.utilities.lambdify import lambdify
```

### Homework to practice the concept of a function and its derivative

1. Given a function of the dependence of the grade on the number of hours spent on studying: $ f(x) = 10x - 5 $.
1. Using Python, calculate the derivative of this function.
2. Use the `matplotlib` library to plot the function and its derivative in the range of values ​​from 0 to 10 hours. Analyze how changing the number of hours of study affects the grade.

```python
x = sp.symbols('x')
f_1 = 10*x + 5    
derivative_f_1 = sp.diff(f_1, x)
print('Derivative of linear function 10x + 5:', derivative_f_1)
```

```python
x_1_val = np.linspace(0, 10, 100)
f_1_val = lambdify(x, f_1, 'numpy')
derivative_f_1_val = [10] * len(x_1_val)
```

```python
plt.figure(figsize=(15, 10))
plt.subplot(1, 3, 1)
plt.plot(x_1_val, f_1_val(x_1_val), label='10x + 5')
plt.title('Linear function 10x + 5')
plt.legend()

plt.subplot(1, 3, 2)    
plt.plot(x_1_val, derivative_f_1_val, label='10')
plt.title('Derivative of linear function 10x + 5')
plt.legend()

plt.subplot(1, 3, 3)
plt.plot(x_1_val, f_1_val(x_1_val), label='10x + 5')
plt.plot(x_1_val, derivative_f_1_val, label='10')
plt.title('Linear function 10x + 5 and its derivative')
plt.legend()
```

### Conclusion of HW 1:

- We have a linearly increasing function.
- The derivative of the function is a constant that is equal to 10
- We can interpret this as each additional hour of study contributes to an increase in the grade by 10 points.
- Therefore, the more hours a student spends on studying, the higher the grade he receives. Which indicates a uniform relationship between study time and the result.

2. Using Python and `SymPy`, calculate the derivatives for the functions $ f(x) = \sin(x) $ and $ f(x) = \ln(x) $. Plot these functions and their derivatives using `matplotlib`.

```python
f_sin = sp.sin(x)
f_ln = sp.log(x)    

derivative_f_sin = sp.diff(f_sin, x)
derivative_f_ln = sp.diff(f_ln, x)
print('Derivative of sin(x):', derivative_f_sin)
print('Derivative of log(x):', derivative_f_ln)
```

```python
x_2_val = np.linspace(0, 10, 100)

f_sin_val = lambdify(x, f_sin, 'numpy')
f_ln_val = lambdify(x, f_ln, 'numpy')

derivative_f_sin_val = lambdify(x, derivative_f_sin, 'numpy')
derivative_f_ln_val = lambdify(x, derivative_f_ln, 'numpy')
```

```python
plt.figure(figsize=(15, 12))
plt.subplot(2, 3, 1)
plt.plot(x_2_val, f_sin_val(x_2_val), label='sin(x)')
plt.title('sin(x)')
plt.legend()
plt.subplot(2, 3, 2)
plt.plot(x_2_val, derivative_f_sin_val(x_2_val), label=derivative_f_sin)
plt.title('Derivative of sin(x)')
plt.legend()
#----
plt.subplot(2, 3, 3)
plt.plot(x_2_val, f_sin_val(x_2_val), label='sin(x)')
plt.plot(x_2_val, derivative_f_sin_val(x_2_val), label=derivative_f_sin)
plt.title('sin(x) and its derivative')
plt.grid()
plt.legend()
#----------------
#----------------
plt.subplot(2, 3, 4)
plt.plot(x_2_val, f_ln_val(x_2_val), label='log(x)')
plt.title('log(x)')
plt.legend()

plt.subplot(2, 3, 5)
plt.plot(x_2_val, derivative_f_ln_val(x_2_val), label=derivative_f_ln)
plt.title('Derivative of log(x)')
plt.legend()

plt.subplot(2, 3, 6)
plt.plot(x_2_val, f_ln_val(x_2_val), label='log(x)')
plt.plot(x_2_val, derivative_f_ln_val(x_2_val), label=derivative_f_ln)
plt.title('log(x) and its derivative')
plt.grid()
plt.legend()
```

### Conclusion of HW 2:

- The function sin(x) and its derivative cos(x)
  - They are periodic functions with a period of 2*pi. And are limited to the values ​​[-1, 1] along *y*
  - The derivative is delayed by pi/2, that is, when sin(x) reaches its maximum or minimum, then cos(x) crosses the OX axis and is equal to 0. That is, at points when the derivative = 0 (and changes sign), the function changes direction (increase to decrease and vice versa). If the derivative has a minus sign along the OX axis, then our function decreases. If the value of the derivative is positive, then the function increases

- The function ln(x) and its derivative 1/x
  - ln(x) exists only for x > 0, which is also true for its derivative
  - ln(x) is a constantly increasing function, but the growth rate decreases with increasing x.
  - The derivative 1/x reflects the rate of change of ln(x). Since the rate of growth of the function decreases, its derivative will also decrease in a similar manner (decreases with increasing x).
  - the derivative also approaches zero, but will never cross it, because our function is monotonically increasing and there is no value for it at which there is a horizontal tangent


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
f_3 = x**2 + 2*x + 1
derivative_f_3 = sp.diff(f_3, x)
print('Derivative of x^2 + 2x + 1:', derivative_f_3)
```

```python
x_3_val = np.linspace(0, 100, 100)
f_3_val = lambdify(x, f_3, 'numpy')
derivative_f_3_val = lambdify(x, derivative_f_3, 'numpy')
```

```python
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.plot(x_3_val, f_3_val(x_3_val), label='x^2 + 2x + 1')
plt.title('x^2 + 2x + 1')
plt.legend()

plt.subplot(1, 3, 2)
plt.plot(x_3_val, derivative_f_3_val(x_3_val), label=derivative_f_3)
plt.title('Derivative of x^2 + 2x + 1')
plt.legend()

plt.subplot(1, 3, 3)
plt.plot(x_2_val, f_3_val(x_2_val), label='x^2 + 2x + 1')
plt.plot(x_2_val, derivative_f_3_val(x_2_val), label=derivative_f_3)
plt.title('x^2 + 2x + 1 and its derivative')
plt.legend()
```
